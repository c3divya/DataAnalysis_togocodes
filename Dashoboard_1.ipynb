```python
colorscale=[[0,'red'],[0.33,'yellow'],[0.66,'blue'],[1,'black']]
trace1=go.Scatter(y=np.random.randn(500),mode='markers',marker=dict(size=16,color=np.random.randn(500),colorscale=colorscale,showscale=True))
data=[trace1]
url1=plot(data,filename='scatter',auto_open=False)
iplot(data,filename='scatter')
```
